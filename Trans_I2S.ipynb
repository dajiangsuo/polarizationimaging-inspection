```python
import tensorflow as tf
from skimage import transform
from skimage import data
import matplotlib.pyplot as plt
import os
import numpy as np
from skimage.color import rgb2gray
import random
```

```python
# Read three types of configuration of polarization images and RGB images
def load_data(data_dir):
    
    #labels = []
    images = []
    file_names = [os.path.join(data_dir,f)
                  for f in os.listdir(data_dir)]
    for f in file_names:
        images.append(data.imread(f))
    return images

train_Polar = "/Users/dajiangsuo/Documents/inspection/"
train_polar_intensity = os.path.join(train_Polar, "Polarization")
train_polar_stokes = os.path.join(train_Polar, "Stokes")
train_polar_params = os.path.join(train_Polar, "Params")
#test_data_dir = os.path.join(ROOT_PATH, "TrafficSigns/Testing")

images_polar_intensity = load_data(train_polar_intensity)
#images_polar_stokes = load_data(train_polar_stokes)
#images_polar_params = load_data(train_polar_params)
```

```python
images_polar_int_array = np.array(images_polar_intensity)
image0 = images_polar_intensity[0]
image0_r = image0.copy()
image0_r[:,:,1]=0
image0_r[:,:,2]=0
image0_g = image0.copy()
image0_g[:,:,0]=0
image0_g[:,:,2]=0
image0_b = image0.copy()
image0_b[:,:,0]=0
image0_b[:,:,1]=0

#Display images in RGB colar
fig, axs = plt.subplots(2,2)
cax_00 = axs[0,0].imshow(image0)
cax_01 = axs[0,1].imshow(image0_r, cmap='Reds')
cax_10 = axs[1,0].imshow(image0_g, cmap='Greens')
cax_11 = axs[1,1].imshow(image0_b, cmap='Blues')

#Display images in grey 
image0_r_2d = image0[:,:,0]
image0_g_2d = image0[:,:,1]
image0_b_2d = image0[:,:,2]
fig1, axs1 = plt.subplots(2,2)
axs1[0,0].imshow(image0)
axs1[0,1].imshow(image0_r_2d, cmap='Greys')
axs1[1,0].imshow(image0_g_2d, cmap='Greys')
axs1[1,1].imshow(image0_b_2d, cmap='Greys')
image0_r_2d.shape
```
